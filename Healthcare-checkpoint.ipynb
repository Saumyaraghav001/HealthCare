```python
plt.figure(figsize=(10,4))
top_10=df['Name'].value_counts().nlargest(10).sort_index()
colors=('yellow','blue','black','red','pink','green','brown','orange','skyblue','darkred')
top_10.plot(kind='bar',color=colors)
plt.xticks(rotation=60)
plt.title("Top 10 doctor")
plt.show()
```

```python
plt.figure(figsize=(10,4))
colors=('yellow','blue','pink','brown','black','green')
top_5=df["Admission Type"].value_counts().nlargest(5).sort_index()
percentages=top_5/(top_5.sum())*100
ax=top_5.plot(kind='barh',width=0.5,color=colors)
for i,v in enumerate(top_5):
    ax.text(v+0.5,i, f'{percentages[i]:.1f}%', color='black', va='center')
plt.xticks(rotation=60)
plt.xlabel("count")
plt.ylabel("Admission Type")
plt.xticks(rotation=60)
plt.yticks(rotation=45)
plt.show()
```
